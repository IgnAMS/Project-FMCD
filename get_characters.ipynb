```python
characters = [
    "fiona_humana", 
    "hada_madrina", 
    "principe_encantador",
    "shrek_humano",
    "shrek_ogro",
    "fiona_ogro",
    "reina_lillian",
    "rey_harold",
    "jengy",
    "burro"
]

videos = ["I_need_a_hero", "Shrek_2_are_we_there", "Great_soup"]
```

### X son en este caso las imagenes, Y son los labels que tienen las imagenes

```python
X = []
Y = []

root_directory = './processing_data/known_faces'

for video_folder in os.listdir(root_directory):
    video_path = os.path.join(root_directory, video_folder)
    for character_folder in os.listdir(video_path):
        character_path = os.path.join(video_path, character_folder)
        label = character_folder
        for image_file in os.listdir(character_path):
            if image_file.endswith(".jpg"):
                image_path = os.path.join(character_path, image_file)
                
                image = cv2.imread(image_path)
                
                # Append the image to X and the label to Y
                X.append(image)
                Y.append(label)
X = np.array(X)
Y = np.array(Y)
```

### Imagenes por personaje

```python
count_character = dict()
for image, label in zip(X, Y):
    if label not in  count_character.keys():
        count_character[label] = 0
    count_character[label] += 1

for character, count in count_character.items():
    print(f"El personaje {character} tiene {count} fotos")
```

## A continuacion 2 fotos por cada personaje

```python
count_character = dict()
for image, label in zip(X, Y):
    if label not in  count_character.keys():
        count_character[label] = 0
    if count_character[label] > 3:
        continue
    count_character[label] += 1
    display(Image(data=cv2.imencode('.jpg', image)[1].tobytes(), format='jpeg'))
    print(f"Label: {label}")
    count_character[label] += 1
```
